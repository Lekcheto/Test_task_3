### Подготовка отчета для создания дашборда

```python
# 1. Отберем все записи по процедуре 'Компьютерная томография органов грудной клетки, 
# брюшной полости и малого таза с контрастированием':

data1 = data[data['PROCEDURE_NAME']  == 
             'Компьютерная томография органов грудной клетки, брюшной полости и малого таза с контрастированием']
display(data1.head())
print(data1.info())
```

<span style="color:blue">Нужные записи отобраны, их осталось 2235 строк.</span>

```python
# 2. Отберем строки в столбце conclusion, по которым есть ключевые слова "recist" и "тип ответа":

data2 = data1[data1['CONCLUSION'].str.lower().str.contains('recist|тип ответа', na = False)]
data2 = data2.reset_index(drop = True)
display(data2.head())
print(data2.info())
```

<span style="color:blue">Нужные строки по заданным ключевым словам отобраны, всего их 179.</span>

```python
# 3. Создадим столбец keyword и запишем отобранные ключевые слова "recist" и "тип ответа":

# создадим пустой столбец keyword
data2['keyword'] = ''

# присвоим строкам в столбце keyword значение "Тип ответа", где "тип ответа" встречается в столбце conclusion
for i in range (data2.shape[0]):
    if 'тип ответа' in data2['CONCLUSION'][i].lower():
        data2['keyword'][i] = 'Тип ответа'
print()

# присвоим строкам в столбце keyword значение "RECIST", где "recist" встречается в столбце conclusion
for j in range (data2.shape[0]):
    if 'recist' in data2['CONCLUSION'][j].lower():
        data2['keyword'][j] = 'RECIST'
print(data2['keyword'].value_counts())
display(data2.head())
```

<span style="color:blue">Итого в столбце keyword 163 строкам присвоено значение "RECIST" и 16 строкам присвоено значение "Тип ответа".</span>

```python
# 4. Создадим столбец estimation_Recist, в который запишем заключение из поля CONCLUSION по ключевым словам:

# создадим список ключевых слов, по которым будем делать поиск
types_answer = ['стабилизация', 'прогрессирование', 'частичный ответ', 'полный ответ']

# создадим список с ответами, которые будет вписывать в столбец estimation_Recist
types_answer_full = ['стабилизация (Stable Disease)', 'прогрессирование (Progressive Disease)', 
                     'частичный ответ (Partial Response)', 'полный ответ (Complete Response)']

# создадим пустой столбец estimation_Recist
data2['estimation_Recist'] = ''

# для строк, где встречаются ключевые слова, впишем ответы
for i in range (data2.shape[0]):
    for j in range(len(types_answer)):
        if types_answer[j] in data2['CONCLUSION'][i].lower():
            data2['estimation_Recist'][i] = types_answer_full[j]
display(data2.head())
data2['estimation_Recist'].value_counts()
```

<span style="color:blue">Столбец estimation_Recist заполнен значениями из поля CONCLUSION, при этом 42 строки остались пустыми.</span>

```python
# 5. Оставим запись в поле «estimation_Recist» пустой, если в CONCLUSION имеются записи, 
# которые указывают на невозможность оценить тип ответа по Recist: 

# создадим список с "отказными" записями
score_unavailable = ['оценка по recist невозможна', 'baseline не указан', 
                     'по критериям recist не выявлены', 'не попадают под критерии recist', 
                     'данные nadir и baseline не указаны']

# сделаем пустыми строки, по которым в столбце conclusion встречаются "отказные" записи
for i in range (data2.shape[0]):
    for j in range(len(score_unavailable)):
        if score_unavailable[j] in data2['CONCLUSION'][i].lower():
            data2['estimation_Recist'][i] = ''
            
display(data2.head())
data2['estimation_Recist'].value_counts()
```

<span style="color:blue">К пустым 42-м строкам добавились еще 8 строк, всего их стало 50.</span>

```python
# 6. Если в CONCLUSION имеется ключевое слово «Recist», но не указан тип ответа, 
# то в поле estimation_Recist укажем «Ошибка описания»:

for i in range (data2.shape[0]):
    if (
    'recist' in data2['CONCLUSION'][i].lower() 
    and 'стабилизация' not in data2['CONCLUSION'][i].lower() 
    and 'прогрессирование' not in data2['CONCLUSION'][i].lower()
    and 'частичный ответ' not in data2['CONCLUSION'][i].lower()
    and 'полный ответ' not in data2['CONCLUSION'][i].lower()
    ):
        data2['estimation_Recist'][i] = 'ошибка описания'
            
display(data2.head())
data2['estimation_Recist'].value_counts()
```

<span style="color:blue">Ошибка описания указана в 42 строках, т.е. в этих строках имеется ключевое слово «Recist», но не указан тип ответа.</span>

```python
# 8. Все ответы в поле «estimation_Recist» должны быть записаны строго в форматах, указанных в п.4 и 6 выше.
```

```python
# 9. Создадим столбец "Date_conclusion" с датой заключения в формате дд.мм.гггг:

data2['STUDY_DATETIME'] = pd.to_datetime(data2['STUDY_DATETIME'])
data2['LAST_VALIDATION_1'] = pd.to_datetime(data2['LAST_VALIDATION_1'])
data2['Date_conclusion'] = data2['STUDY_DATETIME'].dt.strftime('%d.%m.%Y')
data2.head()
```

<span style="color:blue">Столбец Date_conclusion создан в нужном формате.</span>

```python
# 10. Создадим столбец "Time_till_conclusion" и рассчитаем значение как LAST_VALIDATION_1 – STUDY_DATETIME:

# найдем разницу между датами
data2['Time_till_conclusion'] = data2['LAST_VALIDATION_1'] - data2['STUDY_DATETIME']

# переведем разницу между датами в секунды
data2['Time_till_conclusion'] = data2['Time_till_conclusion'].dt.total_seconds()

# создадим функцию для перевода секунд в формат чч.мм.сс
def convert(sec): 
    sec = sec % (24 * 3600) 
    hour = sec // 3600 
    sec %= 3600 
    min = sec // 60 
    sec %= 60 
    return ("%02d.%02d.%02d" % (hour, min, sec)) 

# применим функцию к столбцу 'Time_till_conclusion'
data2['Time_till_conclusion'] = data2['Time_till_conclusion'].apply(convert)
data2.head()
```

<span style="color:blue">Столбец Time_till_conclusion создан в нужном формате.</span>
